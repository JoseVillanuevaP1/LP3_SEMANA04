Formas de Concatenar

```python
#Forma N1
#Dar formato con +
nombre = "Jose Villanueva"
edad = 20
años = str(edad)
print("Buenas Tardes " + nombre + " tienes " + años + " años.")
```

```python
#Forma N2
#Dar formato con .format()
nombre = "Jose Villanueva"
edad = 20
print("Buenas Tardes {} ya se que tienes {}.".format(nombre,edad))
```

```python
#Dar formato a numeros con muchos decimales
resultado = 10/3
print("10/3 es = {} ".format(resultado))
print("10/3 es = {r:1.2f} ".format(r=resultado))
```

```python
#Forma N3
#Dar formato con f
nombre = "Jose Villanueva"
edad = 20
print(f"Buenos dias {nombre}, ya que se que tienes {edad} años")
```

```python
#Dar formato a numeros con muchos decimales con f
resultado = 10/3
print(f"10/3 es = {resultado} ")
print(f"10/3 es = {resultado:1.2f} ")
```
